# Basic QER Recovery Example

This notebook builds a small BGM permutation-invariant code, creates a global symmetric amplitude-damping channel, and optimizes the recovery with the free SCS solver.

```python
from pathlib import Path
import sys

repo_root = Path.cwd().resolve()
if not (repo_root / "src" / "qer").exists():
    repo_root = repo_root.parent
if (repo_root / "src" / "qer").exists():
    sys.path.insert(0, str(repo_root / "src"))

from qer.codewords import bgmcode_piqs
from qer.noisemodel import noisemodel
from qer.optimisation import optimise
```

```python
b, g, m = 3, 3, 1
num_qubits = 2 * b * m + g

rho, logical0, logical1 = bgmcode_piqs(b=b, g=g, m=m)
noise = noisemodel(
    "global symmetric amplitude damping",
    num_qubits=num_qubits,
    gamma=1e-4,
    dt=1.0,
    return_rep="choi",
    dynamics="exact",
)

fidelity = optimise(logical0, logical1, noise, solver="scs")
print(f"Optimized entanglement fidelity: {fidelity:.10f}")
print(f"Optimized entanglement infidelity: {1 - fidelity:.3e}")
```

For larger SDPs, MOSEK can be faster if it is installed and licensed locally: pass `solver="mosek"` to `optimise`.
